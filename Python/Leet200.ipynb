```python
#https://leetcode.ca/all/1730.html
grid=[["X","X","X","X","X","X","X","X"],
      ["X","*","O","X","O","#","O","X"],
      ["X","O","O","X","O","O","X","X"],
      ["X","O","O","O","O","#","O","X"],
      ["X","X","X","X","X","X","X","X"]]
pos = []
def inrange(x,y):
    if x<0 or x>=len(grid):
        return 0
    if y<0 or y>=len(grid[0]):
        return 0
    return 1
for x in range(len(grid)):
    for y in range(len(grid[x])):
        if grid[x][y] == "*":
            pos = [x,y]
            break
    if len(pos):
        break
pos.append(0)
stack=[pos]
visited={}
import math
minn=math.inf
while len(stack):
    x,y,dist=stack.pop(0)
    if grid[x][y] == '#':
        if dist<minn:
            minn=dist
    if (x,y) not in visited:
        visited[(x,y)]=1
        if inrange(x+1,y):
            if grid[x+1][y] in ['O','#']:
                stack.append([x+1,y,dist+1])
        if inrange(x,y+1):
            if grid[x][y+1] in ['O','#']:
                stack.append([x,y+1,dist+1])
        if inrange(x-1,y):
            if grid[x-1][y] in ['O','#']:
                stack.append([x-1,y,dist+1])
        if inrange(x,y-1):
            if grid[x][y-1] in ['O','#']:
                stack.append([x,y-1,dist+1])
print(minn)
```

```python
'''
Input:
["FileSystem","ls","mkdir","addContentToFile","ls","readContentFromFile"]
[[],["/"],["/a/b/c"],["/a/b/c/d","hello"],["/"],["/a/b/c/d"]]

Output:
[null,[],null,null,["a"],"hello"]

'''

class FileSystem:
    fs={}
    def ls(self,path):
        print(self.mkdir(path,listdir=True))
    def mkdir(self,path,insertContent=False,readContent=False,listdir=False):
        tmp = self.fs
        i=0
        folders=path.split('/')
        while len(folders) and folders[0]=='':
            folders.pop(0)
        tot=len(folders)
        if tot==0:
            return tmp
        for sPath in folders:
            if sPath not in tmp :
                if insertContent and i==tot-1:
                    if sPath not in tmp:
                        tmp[sPath] = {}
                    print(i,folders[i])
                    tmp[sPath] = tmp[sPath]+insertContent if len(tmp[sPath]) else insertContent
                elif (readContent or listdir) and i==tot-1:
                    return tmp[sPath]
                else:
                    tmp[sPath] = {}
            else:
                print('path doesnt exist')
            tmp = tmp[sPath]
            i+=1
    def addContentToFile(self,path,content):
        self.mkdir(path , insertContent=content)
        
    def readContentFromFile(self,path):
        print(self.mkdir(path , readContent=True))
        
# Input commands and arguments
commands = ["FileSystem","ls","mkdir","addContentToFile","ls","readContentFromFile"]
arguments = [[],["/"],["/a/b/c"],["/a/b/c/d","hello"],["/"],["/a/b/c/d"]]

# Create an instance of FileSystem
fs = FileSystem()
results = []

# Execute the commands
for command, args in zip(commands, arguments):
    print(fs.fs)
    if command == "FileSystem":
        fs = FileSystem()
        results.append(None)
    elif command == "ls":
        results.append(fs.ls(*args))
    elif command == "mkdir":
        fs.mkdir(*args)
        results.append(None)
    elif command == "addContentToFile":
        fs.addContentToFile(*args)
        results.append(None)
    elif command == "readContentFromFile":
        results.append(fs.readContentFromFile(*args))

# Print the results
print(results)
```

```python
def isSorted(arr):
    for i in range(1,len(arr)):
        if arr[i]<arr[i-1]:
            return False
    return True
def mergeSort(arr):
    if isSorted(arr):
        return arr
    else:
        mid = len(arr)//2
        arr1 = arr[0:mid]
        arr2 = arr[mid:]
        if not isSorted(arr1):
            arr1=mergeSort(arr1)
        if not isSorted(arr2):
            arr2=mergeSort(arr2)
        merged=[]
        i,j=0,0
        while i<len(arr1) and j<len(arr2):
            if arr1[i]<arr2[j]:
                merged.append(arr[i])
                i+=1
            else:
                merged.append(arr[j])
                j+=1
        if i<len(arr1):
            merged.extend(arr1)
        if j<len(arr2):
            merged.extend(arr2)
        return merged 
print(mergeSort([1,4,3,2,5,6,7]))
            
            




```

```python
def binarysearch(low,high,arr,el):
    mid = (low+high) // 2
    if low<=high:
        if arr[mid] == el:
            return mid
        elif arr[mid] > el:
            return binarysearch(low,mid,arr,el)
        else:
            return binarysearch(mid+1,high,arr,el)
arr=[1,2,3,4,5]
print(binarysearch(0,len(arr),arr,4))
            
    
```

```python
words=["abc","ab","bc","b"]
def formTrie(tmp, s):
    tree=tmp
    for x in s:
        if x in tree:
            tree[x]['count']+=1
        else:
            tree[x]={'count':1,'next':{}}
        tree=tree[x]['next']
        print(tree)
    return tmp
def countChars(tmp,s):
    c=0
    tree=tmp
    for x in s:
        if x in tree:
            c+=tree[x]['count']
        else:
            break
        tree=tree[x]['next']
    return c
tree={}
for w in words:
    tree=formTrie(tree,w)
    print(tree)
for w in words:
    print(countChars(tree,w))
```
